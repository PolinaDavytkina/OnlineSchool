# **Арифметические выражения;**

Тут на самом деле все просто в программировании используются такие же действия, как и в математике

```
num1 = 5
num2 = 4

sum = num1 + num2 # сложение
print (sum)
mult = num1 * num2 # умножение
print (mult)
divis = num1 / num2 # деление
print (divis)
subtr = num1 - num2 # вычитание
print (subtr)
expon = num1 ** num2 # возведение в степень
print (expon)
div = num1 // num2 # делочисленное деление
print (div)
mod = num1 % num2 # остаток от деления
print (mod)
```

Арифметические действия не обязательно вкладывать в какую-то переменную, например если они в дальнейшем не будут использоваться в программе

```
num1 = 5
num2 = 4

print(num1 ** num2)
```

# **Условные операторы**

Оператор ветвления if  позволяет выполнить определенный набор инструкций в зависимости от некоторого условия. Возможны следующие варианты использования.

Синтаксис оператора if выглядит так.

if выражение:

    инструкция_1
    инструкция_2
    ...
    инструкция_n
    
После оператора if  записывается выражение. Если это выражение истинно, то выполняются инструкции, определяемые данным оператором. Выражение является истинным, если его результатом является число не равное нулю, непустой объект, либо логическое True. После выражения нужно поставить двоеточие “:”.

```
num = input("Введите число:")
num = int(num)
if num > 5:
  print("Введенное число больше 5")
else:
  print("Введенное число меньше или равно 5")
```

Можно сделать несколько условий

```
num = input("Введите число:")
num = int(num)
if num > 5:
  print("Введенное число больше 5")
elif num == 5:
  print("Введенное число равно 5")
else:
  print("Введенное число меньше 5")
```

# **Операторы цикла**

Цикл while
While - один из самых универсальных циклов в Python, поэтому довольно медленный. Выполняет тело цикла до тех пор, пока условие цикла истинно.

```
i = 5
while i != 0:
  print(i)
  i-=1
```

```
i = 0
while i != 5:
  print(i)
  i+=1
```

Цикл for
Цикл for уже чуточку сложнее, чуть менее универсальный, но выполняется гораздо быстрее цикла while. Этот цикл проходится по любому итерируемому объекту (например строке или списку), и во время каждого прохода выполняет тело цикла.

range() отлично подходит для создания повторяющихся чисел, но это не самый лучший выбор, если вам нужно перебрать данные, которые могут быть зациклены с помощью оператора in.

Есть три способа вызова range():

range(стоп) берет один аргумент

range(старт, стоп) берет два аргумента

range(старт, стоп, шаг) берет три аргумента

Вызывая range() с одним аргументом, вы получите ряд чисел, начинающихся с 0 и включающих каждое число до, но не включая число, которое вы обозначили как конечное (стоп).

```
i = 0

for i in range(5):
  print (i)
```

В фразе "Hello World" давайте посчитаем колличество букв l

```
str1 = "Hello World"
count = 0

for i in str1:
  if i == "l":
    count+=1

print(count)
```

Оператор continue
Оператор continue начинает следующий проход цикла, минуя оставшееся тело цикла (for или while)

А теперь давайте уберем буквы l при печати фразы "Hello World"

```
str1 = "Hello World"

for i in str1:
  if i == "l":
    continue
  print(i, end='')
```

end=' ' - при печати не переводит на новую строку

Оператор break
Оператор break досрочно прерывает цикл.

Давайте выведем все до первой встреченной буквы l

```
str1 = "Hello World"
count = 0

for i in str1:
  if i == "l":
    break
  print(i, end='')
```

Оператор else
Слово else, примененное в цикле for или while, проверяет, был ли произведен выход из цикла инструкцией break, или же "естественным" образом. Блок инструкций внутри else выполнится только в том случае, если выход из цикла произошел без помощи break.

```
str1 = "Hello World"
count = 0

for i in str1:
  if i == "a":
    print("Нашли!")
    break
else:
  print("Буквы 'a' нет в фразе 'Hello World'")
```

# **Функции**

Именные функции, инструкция def
Функция в python - объект, принимающий аргументы и возвращающий значение. Обычно функция определяется с помощью инструкции def.

Инструкция return говорит, что нужно вернуть значение.
Функция может и не заканчиваться инструкцией return, при этом функция вернет значение None

```
def sum(x, y):
  return x + y

a = int(input("Введите число: "))
b = int(input("Введите число: "))

print(sum(a, b))
```

Анонимные функции, инструкция lambda
Анонимные функции могут содержать лишь одно выражение, но и выполняются они быстрее. Анонимные функции создаются с помощью инструкции lambda. Кроме этого, их не обязательно присваивать переменной

lambda функции, в отличие от обычной, не требуется инструкция return, а в остальном, ведет себя точно так же

```
func = lambda x, y: x + y
print(func(1, 2))
print(func('a', 'b'))
```
